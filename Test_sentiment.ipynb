```python
import re
import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
```

```python
corpus=[]

for i in range(0, 1953):
  review = re.sub('[^a-zA-Z]', ' ', dataset['tweet'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
# Loading BoW dictionary
from sklearn.feature_extraction.text import CountVectorizer
import pickle
cvFile='/content/drive/MyDrive/Colab Notebooks/Tweet Sentiment/c1_BoW_Sentiment_Model.pkl'
# cv = CountVectorizer(decode_error="replace", vocabulary=pickle.load(open('./drive/MyDrive/Colab Notebooks/2 Sentiment Analysis (Basic)/3.1 BoW_Sentiment Model.pkl', "rb")))
cv = pickle.load(open(cvFile, "rb"))
```

```python
X_fresh = cv.transform(corpus).toarray()
X_fresh.shape
```

```python
import joblib
classifier = joblib.load('/content/drive/MyDrive/Colab Notebooks/Tweet Sentiment/c2_Classifier_Sentiment_Model')
```

```python
y_pred = classifier.predict(X_fresh)
print(y_pred)
```

```python
len(y_pred)
```

```python
dataset['label'] = y_pred.tolist()
dataset.head()
```

```python
data=dataset[['id','label']]
data
```
