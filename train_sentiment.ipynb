```python
import re
import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
```

```python
corpus=[]

for i in range(0, 7920):
  review = re.sub('[^a-zA-Z]', ' ', dataset['tweet'][i])
  review = ' '.join(re.sub("(\w+:\/\/\S+)", " ", review).split())
  review = ' '.join([w for w in review.split() if len(w)>3])

  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 7000) # Top Frequent 5000
```

```python
dataset.head()
```

```python
X = cv.fit_transform(corpus).toarray()
y = dataset.iloc[:, 1].values
```

```python
y.shape
```

```python
X.shape
```

```python
 # Saving BoW dictionary to later use in prediction
import pickle
bow_path = '/content/drive/MyDrive/Colab Notebooks/Tweet Sentiment/c1_BoW_Sentiment_Model.pkl'
pickle.dump(cv, open(bow_path, "wb"))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB
classifier = MultinomialNB()
classifier.fit(X_train, y_train)
```

```python
# Exporting NB Classifier to later use in prediction
import joblib
joblib.dump(classifier, '/content/drive/MyDrive/Colab Notebooks/Tweet Sentiment/c2_Classifier_Sentiment_Model') 
```

```python
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)

accuracy_score(y_test, y_pred)
```
